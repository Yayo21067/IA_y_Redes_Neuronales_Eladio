```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv")
```

```python
print("Primeras 5 filas")
print(df.head(5))
print("\nTipos de datos")
print(df.info())
```

```python
print("Valores Nulos ")
print(df.isnull().sum())
```

```python
print("Valores Nulos ")
print(df.isnull().sum())
mediana_hp = df['horsepower'].median()
df['horsepower'] = df['horsepower'].fillna(mediana_hp)
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['acceleration'], bins=30, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribución de la Aceleración')
plt.xlabel('Aceleración (m/s²)')
plt.ylabel('Cantidad de Vehículos')
plt.grid(axis='y',alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['displacement'], df['horsepower'], alpha=0.5, color='darkorange')
plt.title('Relación: Desplazamiento del Motor vs Caballos de Fuerza')
plt.xlabel('Desplazamiento del Motor')
plt.ylabel('Caballos de Fuerza')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
resumen_origen = df.groupby('origin')['mpg'].mean()
resumen_origen.plot(kind='bar', color=['blue', 'red', 'green'])
plt.title('Rendimiento Promedio (MPG) por Origen')
plt.xlabel('País de Origen')
plt.ylabel('Promedio de Millas por Galón')
plt.xticks(rotation=0)
plt.show()
```
